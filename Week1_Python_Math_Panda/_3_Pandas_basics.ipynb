# Task
Concatenate a DataFrame with itself after splitting it into two halves.

## Determine the split point

### Subtask:
Calculate the index at which to split the DataFrame into two halves.


**Reasoning**:
Calculate the split point of the DataFrame by getting the number of rows and dividing by 2.



```python
num_rows = df.shape[0]
split_point = num_rows // 2
```

## Split the dataframe

### Subtask:
Create two new DataFrames, each representing one half of the original DataFrame.


**Reasoning**:
Create two new DataFrames representing the first and second halves of the original DataFrame using the calculated split point.



```python
df_first_half = df.iloc[:split_point]
df_second_half = df.iloc[split_point:]
display(df_first_half.head())
display(df_second_half.head())
```

## Concatenate the dataframes

### Subtask:
Concatenate the two halves of the DataFrame (`df_first_half` and `df_second_half`) back into a single DataFrame.


**Reasoning**:
Concatenate the two halves of the DataFrame using pd.concat().



```python
df_concatenated = pd.concat([df_first_half, df_second_half])
display(df_concatenated.head())
display(df_concatenated.tail())
```

## Verify the concatenation

### Subtask:
Verify the concatenation by displaying the head and shape of the concatenated DataFrame to ensure it matches the original DataFrame.


**Reasoning**:
Display the head and shape of the concatenated DataFrame to verify the concatenation.



```python
display(df_concatenated.head())
display(df_concatenated.shape)
```
