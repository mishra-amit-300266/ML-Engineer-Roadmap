```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

**Line Plot**

Used to visualize the trend of a continuous variable over a range (like time or angle).

In this example, we plotted a sine wave using np.sin(x) to see the smooth periodic behavior.


```python
# Generating 100 values between 0 and 10
x = np.linspace(0, 10, 100)

# Taking sine of those values
y = np.sin(x)

# Creating a line plot
plt.plot(x, y,color='red')

# Adding title and labels
plt.title("Sine Wave")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Adding grid lines
plt.grid(True)

# Displaying the plot
plt.show()
```

**Scatter Plot**

Used to analyze the relationship between two numerical variables.

Each point represents an observation. Helpful for spotting clusters, trends, or outliers.

```python
# Generating 50 random x and y values between 0 and 1
x = np.random.rand(50)
y = np.random.rand(50)

# Creating a scatter plot
plt.scatter(x, y, color='green')
plt.title("Random Scatter")
plt.show()
```

**Bar Chart**

Used to compare quantities across different categories.

Each bar's height represents the value for that category. Ideal for visualizing categorical data.

```python
# Categorical data
categories = ['A', 'B', 'C']
values = [10, 24, 36]

# Creating a bar chart
plt.bar(categories, values, color='purple')
plt.title("Bar Chart Example")
plt.show()
```

**Histogram**

Used to understand the distribution of a single numerical variable.

Shows frequency of data within specified bins. Helpful for detecting skewness or normality.

```python
# Generating 1000 random values from a normal distribution
data = np.random.randn(1000)

# Creating a histogram with 30 bins
plt.hist(data, bins=30, color='skyblue')
plt.title("Histogram Example")
plt.show()
```

**Subplots**

Used to understand the distribution of a single numerical variable.

Shows frequency of data within specified bins. Helpful for detecting skewness or normality.

```python
# Generating x values and computing sine and cosine
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Setting the overall figure size
plt.figure(figsize=(10, 4))

# First subplot (Sine)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.plot(x, y1)
plt.title("Sine")

# Second subplot (Cosine)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.plot(x, y2)
plt.title("Cosine")

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Importing seaborn and pandas
import seaborn as sns
import pandas as pd
```

```python
# Loading seaborn's built-in Titanic dataset
df = sns.load_dataset("titanic")

# Displaying the first few rows
df.head()
```

**Histogram with kde**

A histogram combined with KDE (Kernel Density Estimation) shows both raw frequency and smoothed distribution.

Great for understanding the shape, skewness, and spread of a variable.

```python
# Plotting distribution of age with histogram + KDE (line showing density)
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

**BoxPlot**

Shows summary statistics: median, quartiles, and outliers of a numerical variable, grouped by categories.

Ideal for comparing distributions (e.g., fares across passenger classes).

```python
# Boxplot shows median, quartiles, and outliers
sns.boxplot(x="pclass", y="fare", data=df)
plt.title("Fare by Passenger Class")
plt.show()
```

**CountPlot**

Bar chart that shows the count of occurrences of categories.

Useful for checking how many people survived vs didnt, or how many passengers were in each class.

```python
# Count of survivors (0 = No, 1 = Yes)
sns.countplot(x="survived", data=df)
plt.title("Survival Count")
plt.show()
```

**HeatMap**

Displays correlation (or other matrix-style data) as a color-coded grid.

Each cell shows the strength of the relationship between two variables. Helps in feature selection.

```python
# Calculating correlation between numeric columns
corr = df.corr(numeric_only=True)

# Displaying heatmap of correlations
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

**PairPlot**

Plots pairwise relationships among several numeric features. You can use the 'hue' parameter to color-code by a category (e.g., survival).

Useful to spot clusters, patterns, or relationships across multiple variables.

```python
# Visualizing pairwise relationships between features, color-coded by survival
sns.pairplot(df[["age", "fare", "pclass", "survived"]].dropna(), hue="survived")
plt.show()
```
