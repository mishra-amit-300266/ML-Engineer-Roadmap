1. Introduction to Pandas

Pandas is a Python library used for data analysis and manipulation. It provides powerful tools to work with tabular data (like Excel or SQL tables).


```python
import pandas as pd
import numpy as np
```

2. Core Data Structures

i. Series (1D)

ii. DataFrame (2D)

```python
s = pd.Series([1,2,3])
print(s)
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'San Francisco', 'Los Angeles']
}
df = pd.DataFrame(data)
print(df)
```

4. Exploring Data

```python
df.head()           # First 5 rows
df.tail()           # Last 5 rows
df.shape            # (rows, columns)
df.columns          # Column names
df.dtypes           # Data types
df.info()           # Summary
df.describe()       # Statistics
df['Age'].value_counts()
```

5. Indexing and Filtering

```python
#Selecting Columns
df['Age']
```

```python
#Row by Label vs Position
df.loc[0]     #by label/index
df.iloc[0]    #by position
```

```python
#Boolean Filtering
df[df['Age'] > 30]
```

6. Data Cleaning

```python
#Missing Values
df.isnull().sum()
df.fillna(0)          # Replace NaNs
df.dropna()           # Remove rows with NaNs
```

```python
#Changing Dtypes
df['Age'] = df['Age'].astype(int)
```

```python
#Renaming
df.rename(columns={'Age': 'Age_Years'}, inplace = True)
df
```

```python
#Replacing
df['City'].replace({'New York': 'NYC'}, inplace = True)
df
```

7. Data Operations

```python
#Sorting
df.sort_values(by = 'Age_Years', ascending = False)
```

```python
#Grouping and Aggregation
df.groupby('City')['Age_Years'].mean()
```
